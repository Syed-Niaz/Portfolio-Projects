```python
#importing libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
#import the data and view the data

df=pd.read_csv(r"D:\PROJECTS\Alex Projects\Python\movies.csv")
df
```

```python
#Check df to see if there are any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#Drop the rows with missing data

df.dropna(axis = 0 , inplace = True)
```

```python
#see datatype of each column

df.dtypes
```

```python
#change 'budget' & 'gross' column dtype from float64 to int64

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
#Display max rows

pd.set_option('display.max_rows', None)
```

```python
#reset

pd.reset_option('all')
```

```python
#dropping duplicate values

df.drop_duplicates()
```

```python
#scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')

plt.show()
```

```python
#Order by 'gross'

df.sort_values('gross', ascending= True, inplace= False)
```

```python
#plot budget vs gross with seaborn

#there is a positive correlation between 'budget' and 'gross'

sns.regplot(x= 'budget', y= 'gross', data = df, scatter_kws= {"color": "blue"}, line_kws= {"color": "red"})
```

```python
#see correlation

df.corr(method='pearson')
```

```python
sns.heatmap(data = df.corr(method='pearson'), annot= True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
#Giving every object dtype a numeric identity so that we can see the correlation between everything

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
#Creating a heatmap visualization to see correlation of everything

correlation_matrix = df_numerized.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot= True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```
